```python
#1
import sqlite3
```

```python
#2 to create database intern_training_db
conn = sqlite3.connect("intern_training_db.db")
cursor = conn.cursor()
```

```python
#3
#SQLite does not use schemas explicitly (unlike MySQL).
```

```python
#4
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    email TEXT,
    age INTEGER
)
""")
```

```python
#5
cursor.execute("""
INSERT INTO students (name, email, age) VALUES
('Amit Sharma', 'amit@gmail.com', 21),
('Riya Patel', 'riya@gmail.com', 22),
('Karan Verma', 'karan@gmail.com', 20),
('Sneha Rao', 'sneha@gmail.com', 23),
('Rahul Mehta', 'rahul@gmail.com', 24)
""")

conn.commit() #Without commit â†’ data is lost after program ends
```

```python
#6
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall() #returns all rows

for row in rows:
    print(row) #to print each record
```

```python
#7
cursor.execute("SELECT name FROM students")
for row in cursor.fetchall():
    print(row)
```
